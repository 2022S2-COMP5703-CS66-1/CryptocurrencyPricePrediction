```python
# 将hourly的格式改一下
a=df_new['hour'].tolist()
b=df_new['AMORPM'].tolist()

c=[]
for i in range(0,len(a)):
    if b[i]=='AM'and a[i]=='12:00:00':k='00:00:00'
    elif b[i]=='PM'and a[i]=='1:00:00':k='13:00:00'
    elif b[i]=='PM'and a[i]=='2:00:00':k='14:00:00'
    elif b[i]=='PM'and a[i]=='3:00:00':k='15:00:00'
    elif b[i]=='PM'and a[i]=='4:00:00':k='16:00:00'
    elif b[i]=='PM'and a[i]=='5:00:00':k='17:00:00'
    elif b[i]=='PM'and a[i]=='6:00:00':k='18:00:00'
    elif b[i]=='PM'and a[i]=='7:00:00':k='19:00:00'
    elif b[i]=='PM'and a[i]=='8:00:00':k='20:00:00'
    elif b[i]=='PM'and a[i]=='9:00:00':k='21:00:00'
    elif b[i]=='PM'and a[i]=='10:00:00':k='22:00:00'
    elif b[i]=='PM'and a[i]=='11:00:00':k='23:00:00'
    else:k=a[i]
    c.append(k)
```
